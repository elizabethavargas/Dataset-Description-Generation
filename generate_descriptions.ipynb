```python
llama_generator = HFGenerator("unsloth/Meta-Llama-3.1-8B-Instruct")
```

### Generate Non Few Shot Descriptions

```python
import pandas as pd

sampled_datasets = pd.read_pickle('/content/sampled_datasets.pkl')

records = []
for sample in sampled_datasets:
    dataset_id = sample["dataset_id"]

    try:
        description = llama_generator.generate_description(
            sample,
            user_focused=True,
            few_shot=False,
            temperature=0.0
        )
    except Exception as e:
        print(f"❌ Failed on dataset_id: {dataset_id} — {e}")
        description = None

    records.append({
        "dataset_id": dataset_id,
        "description": description
    })

descriptions_df = pd.DataFrame(records)
descriptions_df.to_pickle("/content/dataset_descriptions.pkl")
descriptions_df.head()
```

## Generate Few Shot Descriptions

```python
import pandas as pd

sampled_datasets = pd.read_pickle('/content/sampled_datasets.pkl')

records = []
for sample in sampled_datasets:
    dataset_id = sample["dataset_id"]

    try:
        description = llama_generator.generate_description(
            sample,
            user_focused=True,
            few_shot=True,
            temperature=0.0
        )
    except Exception as e:
        print(f"❌ Failed on dataset_id: {dataset_id} — {e}")
        description = None

    records.append({
        "dataset_id": dataset_id,
        "description": description
    })

descriptions_df = pd.DataFrame(records)
descriptions_df.to_pickle("/content/dataset_descriptions_few_shot.pkl")
descriptions_df.head()
```
